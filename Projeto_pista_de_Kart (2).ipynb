```python
import numpy as np
import pandas as pd
```

```python
numero_corredores = 6
numero_voltas = 10

matriz = np.random.uniform(20, 60, size=(numero_corredores, numero_voltas))
dados = pd.DataFrame(matriz)
nomes_corredores = ["Felipe Massa", "Diniz", "Alonso", "Jean", "Gustavo", "Mario"]
dados.columns = ['Volta 1', 'Volta 2', 'Volta 3', 'Volta 4', 'Volta 5',
                 'Volta 6', 'Volta 7', 'Volta 8', 'Volta 9', 'Volta 10']
dados.index = nomes_corredores
dados.head()
```

```python
min_tempo = dados.min().idxmin()
indice_min = dados.values.argmin()
media_tempo = dados.mean().idxmin()
soma_tempo = dados.sum(axis=1)
campeao = soma_tempo.idxmin()
piloto_media = dados[media_tempo].idxmin()
piloto_min = dados[min_tempo].idxmin()
piloto_ind = dados.stack().idxmin()[0]

print(f'esse é a soma do tempo:')
print(soma_tempo)
print(f'A melhor {min_tempo} {piloto_min} ')
print(f'Indice da melhor volta: {indice_min} {piloto_ind} ')
print(f'A volta média mais rápida foi: {media_tempo} {piloto_media}')
print(f'A soma de todos os tempos: ')
print(soma_tempo)
print(f'O campeão foi o corredor {campeao}')
dados.describe()
```
